## Section 3 : Handling Missing Values
1. Display total null values in each column.
2. Drop rows where embarked is null.
3. Fill missing values in age using:
- Mean(create new column:age_mean_filled)
- Median(create new column:age_median_filled)
4. Exlpain which method(mean or median) is more apporopriate and why.
5. Verify that missing values are handled correctly.

```python
df.isnull().sum()
```

```python
df.dropna(subset=['embarked'],inplace=True)
df.isnull().sum()
```

```python
df.shape
```

```python
df['Age_mean_filled']=df['Age'].fillna(df['Age'].mean())
```

```python
df['Age_median_filled']=df['Age'].fillna(df['Age'].median())
```

```python
df.head(2)
```

## Median is more appropriate
- as we have seen the data of age contains outliers
- mean gets effected by outliers but median get less effected by outliersso it is more appropriate
